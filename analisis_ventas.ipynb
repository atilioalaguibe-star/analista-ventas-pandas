```python
import pandas as pd
```

```python
ventas = { "Producto": ["Laptop","Mouse", "Teclado", "Monitor"],
           "Cantidad": [2,10,5,3],
           "Precio": [1200,25,60,300]}
```

```python
df = pd.DataFrame(ventas)
df
```

```python
df ["ingresos"] = df["Cantidad"] * df ["Precio"]
```

```python
df
```

```python
df.sort_values(by="ingresos",ascending=False)
```

```python
df ["Cantidad"] = pd.to_numeric(df["Cantidad"])
df ["Precio"] = pd.to_numeric(df["Precio"])
```

```python
df["ingresos"] = df["Cantidad"] * df["Precio"]
```

```python
df.dtypes
```

```python
df.groupby("Producto") ["ingresos"].sum().sort_values(ascending=False)
```

```python
df["ingresos"].sum()
```

```python
df ["ingresos"].mean()
```

```python
df.groupby("Producto") ["ingresos"].sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

df.groupby("Producto") ["ingresos"].sum().sort_values().plot(kind="barh")
plt.title("Ingresos por Producto")
plt.show()
```
